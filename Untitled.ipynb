```python
def read_unlabeled_image_list(image_list_file):

    root = image_list_file
#     pattern = exp
    images = []
    for path, subdirs, files in os.walk(root):
        for name in files:
            if name.endswith(('.jpg', '.jpeg', '.gif', '.png')):
                images.append(path +"/"+ name)

    return images
```

```python
root = 'segment/'
all_images = read_unlabeled_image_list(root)
```

```python
annotation.shape
```

```python
i = 0
while(True):
    try:
        seg_path = rand.choice(all_images)
        im_path = seg_path.replace('segment', 'images')
        
        
        encoded = np.array(Image.open(seg_path))
        photo   = Image.open(im_path)

        annotation = np.bitwise_or(np.bitwise_or(
            encoded[:, :, 0].astype(np.uint32),
            encoded[:, :, 1].astype(np.uint32) << 8),
            encoded[:, :, 2].astype(np.uint32) << 16)

        plt.figure(figsize=(20,10))
        
        plt.subplot(122)
        plt.imshow(photo)
        
        plt.subplot(121)
        plt.imshow(annotation)
        
        
        break
    except:
        pass
    
print(i)
```

```python

src_im =  im_path
src_seg = seg_path 
dst_im =  os.path.join(os.getcwd(), "mustafa_new_dataset\\" + im_path)
dst_seg = os.path.join(os.getcwd(), "mustafa_new_dataset\\" + seg_path)


traget_path_im = "mustafa_new_dataset/" + "/".join(im_path.split("/")[:-1])
traget_path_seg = "mustafa_new_dataset/" + "/".join(seg_path.split("/")[:-1])


if not os.path.exists(traget_path_im):
    os.makedirs(traget_path_im)
    
if not os.path.exists(traget_path_seg):
    os.makedirs(traget_path_seg)
    

    
copyfile(src_im, dst_im)
copyfile(src_seg, dst_seg)

```
